```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
feature_data = pd.read_csv("custom_more_data.csv",index_col=[0,1,2],header=[0,1,2])
```

```python
feature_data.index.rename(['index', 'track_id','gerne'], inplace=True)
```

```python
mel_spec = pd.read_csv("custom_more_data_mel_spec.csv",index_col=[0,1,2],header=[0,1,2])
```

```python

mel_spec.index.rename(['index', 'track_id','gerne'], inplace=True)
```

```python
feature_data.reset_index()
mel_spec.reset_index()
```

```python
feature_data
```

```python
all_data = feature_data.reset_index().merge(mel_spec.reset_index(),on="track_id")
```

```python
all_data
```

```python
all_data.drop("index_x",inplace=True,axis=1)
all_data.drop("gerne_y",inplace=True,axis=1)
all_data.drop("index_y",inplace=True,axis=1)
```

```python
all_data.set_index(["track_id"],inplace=True)
```

```python
# all_data.index.rename(['track_id'], inplace=True)
```

```python
all_data
```

```python
only_pop = all_data[all_data['gerne_x']=="pop"]
```

```python
only_pop['gerne_x'] = only_pop['gerne_x'].replace({"pop":"Pop"})
```

```python
only_pop = only_pop.rename(columns={'gerne_x': 'genre_top'})
```

```python
only_pop.to_csv("new_data_only_pop.csv")
```

```python
all_data.reset_index(1,inplace=True)
```

```python
all_data.dropna(inplace=True)
```

```python
all_data
```

```python
X = all_data.drop('genre_top',axis=1)
y = all_data['genre_top']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = RandomForestClassifier(n_estimators = 200)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```
