```python
import pandas as pd
```

```python
tef = pd.read_csv("track_echonest_feature_clean.csv",index_col=[0,1],header=[0,1,2])
```

```python
tef
```

```python
tef.reset_index(inplace=True)
tefCol = tef.columns
```

```python
tefCol
```

```python
tef.drop(('echonest','acousticness','Unnamed: 2_level_2'),axis=1,inplace=True)
tef.drop(('echonest','danceability','Unnamed: 3_level_2'),axis=1,inplace=True)
tef.drop(('echonest','instrumentalness','Unnamed: 5_level_2'),axis=1,inplace=True)
tef.drop(('echonest','liveness','Unnamed: 6_level_2'),axis=1,inplace=True)
tef.drop(('echonest','speechiness','Unnamed: 7_level_2'),axis=1,inplace=True)
tef.drop(('echonest','valence','Unnamed: 9_level_2'),axis=1,inplace=True)
```

```python
tef
```

```python
selected_genre = tef[tef["genre_top"].isin([15,12,10,17,25,1,21,32,107,76,41,27,18,42,66])]
```

```python
selected_genre
```

```python
X = tef["mfcc"]
y = tef["genre_top"]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(sampling_strategy='auto', random_state=42)
X_train_resampled, y_train_resampled = rus.fit_resample(X_train, y_train)
```

```python
model = RandomForestClassifier(n_estimators=100)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = pd.DataFrame(y_pred)
```

```python
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred,average='micro')
recall = recall_score(y_test, y_pred,average='micro')
f1 = f1_score(y_test, y_pred,average='micro')
confusion = confusion_matrix(y_test, y_pred)
mean_squared_error = mean_squared_error(y_test, y_pred)
r2_score = r2_score(y_test, y_pred)


# Print metricsµ
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
print(f"Confusion Matrix:\n{confusion}")
print(f"Mean Squared Error: {mean_squared_error}")
print(f"R2 Score: {r2_score}")
```
