```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
mfcc= pd.read_csv("mfcc_clean.csv",index_col=[0])
```

```python
track_genre_filled = pd.read_csv("../Notebooks/track_genre_mapping.csv",index_col=[0])
```

```python
all_data = track_genre_filled.merge(mfcc,left_index=True,right_index=True)
```

```python
selected_model_data = all_data[all_data['genre_top'].isin(["Rock","Electronic","Pop","Hip-Hop","Folk"])]
```

```python
X = selected_model_data.drop("genre_top",axis=1)
y = selected_model_data['genre_top']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='auto', random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
model = RandomForestClassifier(n_estimators = 300)
```

```python
model.fit(X_train_resampled,y_train_resampled)
```

```python
y_pred=model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```
