```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv("custom_more_data.csv",index_col=[0,1,2],header=[0,1,2])
```

```python
data.index.rename(["index","track_id","genre_top"],inplace=True)
```

```python
data.reset_index([0,2],inplace=True)
```

```python
data.drop("index",axis=1,inplace=True)
```

```python
data
```

```python
mel_spec = pd.read_csv("custom_more_data_mel_spec.csv",index_col=[0,1,2],header=[0,1,2])
```

```python
mel_spec.index.rename(["index","track_id","genre_top"],inplace=True)
```

```python
mel_spec.reset_index([0,2],inplace=True)
```

```python
mel_spec.drop(["index","genre_top"],inplace=True,axis=True)
```

```python
all_gtzan = data.merge(mel_spec,left_index=True,right_index=True)
```

```python
all_gtzan
```

```python
X = all_gtzan.drop("genre_top",axis=1)
y = all_gtzan['genre_top']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = RandomForestClassifier(n_estimators = 300,max_depth=None)
```

```python
model.fit(x_train,y_train)
```

```python
importances = model.feature_importances_
feature_names = x_train.columns  # Or list of your feature names

# Create a DataFrame
feat_importances = pd.DataFrame({'feature': feature_names, 'importance': importances})

# Sort by importance
feat_importances = feat_importances.sort_values(by='importance', ascending=False)
```

```python
# import matplotlib.pyplot as plt

# # Plot top 50 features
# top_n = 50

# plt.figure(figsize=(10, 12))
# plt.barh(feat_importances['feature'][:top_n], feat_importances['importance'][:top_n])
# plt.gca().invert_yaxis()  # Most important on top
# plt.xlabel('Importance')
# plt.title(f'Top {top_n} Feature Importances')
# plt.show()
```

```python
top_n = 250
```

```python
top_features = feat_importances['feature'][:top_n].tolist()

# Filter datasets
X_train_selected = x_train[top_features]
X_test_selected = x_test[top_features]
```

```python
model.fit(X_train_selected,y_train)
```

```python
y_pred=model.predict(X_test_selected)
```

```python
print(classification_report(y_test, y_pred))
```
